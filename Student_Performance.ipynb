```python
# without reduce (using loop)
numbers = [1, 2, 3, 4]

total = 0
for n in numbers:
    total += n

print(total)


# with reduce (using function)
from functools import reduce

def add(a, b):
    return a + b

numbers = [1, 2, 3, 4]
result = reduce(add, numbers)

print(result)


# with reduce + lambda
numbers = [1, 2, 3, 4]
result = reduce(lambda a, b: a + b, numbers)

print(result)
```
